```python
#Import the libraries
from sklearn.cluster import DBSCAN
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import normalize
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
airline = pd.read_csv("EastWestAirlines.csv")
print(airline.head())
```

```python
airline.info()
```

```python
airline=airline.drop(['ID#'],axis=1)
airline
```

```python
#Normalize function
airline_norm=pd.DataFrame(normalize(airline),columns=airline.columns)
airline_norm
```

```python
#Dendogram
plt.figure(figsize=(10,7))
dendograms=sch.dendrogram(sch.linkage(airline_norm,'complete'))
```

```python
#create clustering
hclusters= AgglomerativeClustering(n_clusters=5,affinity='euclidean', linkage='complete')
hclusters
```

```python
y=pd.DataFrame(hclusters.fit_predict(airline_norm),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
# Adding clusters to dataset
airline['clustersid']=hclusters.labels_
airline
```

```python
airline.groupby(['clustersid']).agg(['mean']).reset_index()
```

```python
# Plot Clusters
plt.figure(figsize=(9, 6))  
plt.scatter(airline['clustersid'],airline['Balance'], c=hclusters.labels_) 
```

# Perform Clustering(Hierarchical, Kmeans & DBSCAN) for the crime data and identify the number of clusters formed and draw inferences.

Data Description:
Murder -- Muder rates in different places of United States

Assualt- Assualt rate in different places of United States

UrbanPop - urban population in different places of United States

Rape - Rape rate in different places of United States


```python
from sklearn.cluster import DBSCAN
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import normalize
import scipy.cluster.hierarchy as sch
```

```python
crime=pd.read_csv('crime_data.csv')
crime
```

```python
crime.info()
```

```python
crime.drop(['Unnamed: 0'],axis=1,inplace=True)
crime
```

```python
# Normalize heterogenous numerical data using standard scalar fit transform to dataset
crime_norm=StandardScaler().fit_transform(crime)
crime_norm
```

```python
#Dendogram
plt.figure(figsize=(10,7))
dendograms=sch.dendrogram(sch.linkage(crime_norm,'average'))
```

```python
# DBSCAN Clustering
dbscan=DBSCAN(eps=1,min_samples=4)
dbscan.fit(crime_norm)
```

```python
#Noisy samples are given the label -1.
dbscan.labels_
```

```python
crime['clusters']=dbscan.labels_
crime
```

```python
crime.groupby('clusters').agg(['mean']).reset_index()
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))  
plt.scatter(crime['clusters'],crime['UrbanPop'], c=dbscan.labels_) 
```
